```python
#Pipeline

#Sentiment Analysis

!pip install transformers

from transformers import pipeline
classifier=pipeline("sentiment-analysis")
one=classifier("I'm feeling good")  #positive sentence
two=classifier("I'm feeling bad")   #negative sentence
three=classifier("I'm not feeling good")  #negative of positive sentence which is essentially a negative sentence
four=classifier("I'm not feeling bad")    #negative of negative sentence which is essentially a positive sentence
print(one,two,three,four)
```
